# Analyze reproducibility
Different methods can be used to evaluate accuracy or reproducibility of a clustering algorithm. Some examples are shown in Scikit-learns [Clustering performance evaluation](https://scikit-learn.org/stable/modules/clustering.html#clustering-performance-evaluation).

Methods that could be applied to this data: 
* **Adjuster rand index:** This does technically require a set of known true cluster labels, which we do not have, but could be used to evaluate all clusters against the first round of clustering (setting the first round as the "ground truth").
* **Silhouette Coefficient:** Does not require known true cluster labels. Is applied once for each clustering and estiamtes how well defined the clusters are.

```python
#We still don't have a measurement of reproducibility for this compound. 
compound_dict = {'jq1': '(+)-JQ1', 
                 'ibet151': 'I-BET151', 
                 'panobinostat': 'CBK277990',
                 'vemurafenib': 'CBK277976',
                 'crizotinib': 'CBK277930'
                }

for compound in compound_dict.keys():
    #Attempt at measuring repeatability:
        #Create a matrix with all samples as rows and a column for each compound we are interested in. 
    data_repeat = pd.DataFrame.from_dict({'sc3s_100': [0]*10941,
                                           'sc3s_110': [0]*10941,
                                           'sc3s_120': [0]*10941,
                                           'sc3s_150': [0]*10941,
                                           'sc3s_180': [0]*10941
                                          })
        # Each time a sample is included in the cluster for that compound, +1.
    longest_clust_len = 0
    longest_clust = 'sc3s_100'
    for k in range(10):
        data_temp = pd.read_parquet('/share/data/analyses/camilla/cp_tpp/processed_data/pca_specs_all_withClusters_'+str(12+k)+'.parquet')
        for column in data_repeat.columns:
            cluster_no = list(data_temp[data_temp["cbkid"] == compound_dict[compound]][column])[0]
            for i in data_repeat.index:
                if data_temp.loc[i,column] == cluster_no:
                    data_repeat.loc[i,column] += 1
            clust_len = len(data_repeat.loc[data_repeat[column] > 0,column])
            if clust_len > longest_clust_len:
                longest_clust_len = clust_len
                longest_clust = column
    
        # Heatmap?
    ax = sns.clustermap(data_repeat.loc[data_repeat[longest_clust] > 0,], 
                   col_cluster=False, 
                   row_cluster=True, 
                   cmap="vlag", 
                   center=0, 
                   yticklabels=False)
    
    #ax.set(xlabel="k in k-mean clustering", ylabel="Sample")
    plt.savefig("../../figures/figures_from_template/heatmap_sc3s_countSampleInCluster_"+compound+".svg")

    #Calculate percentage of each cluster which occur in 9 of 10 replicates 
    data_percent = pd.DataFrame.from_dict({'sc3s_100': [0.0]*10,
                                           'sc3s_110': [0.0]*10,
                                           'sc3s_120': [0.0]*10,
                                           'sc3s_150': [0.0]*10,
                                           'sc3s_180': [0.0]*10
                                          })
    data_clusterSize = pd.DataFrame.from_dict({'sc3s_100': [0.0]*10,
                                           'sc3s_110': [0.0]*10,
                                           'sc3s_120': [0.0]*10,
                                           'sc3s_150': [0.0]*10,
                                           'sc3s_180': [0.0]*10
                                          })
    
    for k in range(10):
        data_temp = pd.read_parquet('/share/data/analyses/camilla/cp_tpp/processed_data/pca_specs_all_withClusters_'+str(12+k)+'.parquet')
        for column in data_repeat.columns:
            cluster_no = list(data_temp[data_temp["cbkid"] == compound_dict[compound]][column])[0]
            sample_total_cluster = 0.0
            for i in data_repeat.index:
                if data_temp.loc[i,column] == cluster_no:
                    sample_total_cluster += 1.0
                    #Check if sample i occur in 9 of 10 replicates
                    if data_repeat.loc[i,column] >= 9:
                        data_percent.loc[k,column] += 1.0
            data_percent.loc[k,column] = data_percent.loc[k,column]/sample_total_cluster*100
            data_clusterSize.loc[k,column] = sample_total_cluster

    #Save tables for data_percent and data_clusterSize
    data_percent.to_latex("../../tables/sc3s_reproducibility_percentSampleIn9of10Clusters_600rep_"+compound+".txt")
    data_clusterSize.to_latex("../../tables/sc3s_reproducibility_clusterSize_600rep_"+compound+".txt")
```

```python
compound_dict = {'jq1': '(+)-JQ1', 
                 'ibet151': 'I-BET151', 
                 'panobinostat': 'CBK277990',
                 'vemurafenib': 'CBK277976',
                 'crizotinib': 'CBK277930'
                }

for compound in compound_dict.keys():
    #Attempt at measuring repeatability:
        #Create a matrix with all samples as rows and a column for each compound we are interested in. 
    data_repeat = pd.DataFrame.from_dict({'sc3s_100': [0]*10941,
                                           'sc3s_110': [0]*10941,
                                           'sc3s_120': [0]*10941,
                                           'sc3s_150': [0]*10941,
                                           'sc3s_180': [0]*10941
                                          })
        # Each time a sample is included in the cluster for that compound, +1.
    longest_clust_len = 0
    longest_clust = 'sc3s_100'
    for k in range(10):
        data_temp = pd.read_parquet('/share/data/analyses/camilla/cp_tpp/processed_data/pca_specs_all_withClusters_2000runs'+str(k)+'.parquet')
        for column in data_repeat.columns:
            cluster_no = list(data_temp[data_temp["cbkid"] == compound_dict[compound]][column])[0]
            for i in data_repeat.index:
                if data_temp.loc[i,column] == cluster_no:
                    data_repeat.loc[i,column] += 1
            clust_len = len(data_repeat.loc[data_repeat[column] > 0,column])
            if clust_len > longest_clust_len:
                longest_clust_len = clust_len
                longest_clust = column
    
        # Heatmap?
    print(compound)
    ax = sns.clustermap(data_repeat.loc[data_repeat[longest_clust] > 0,], 
                   col_cluster=False, 
                   row_cluster=True, 
                   cmap="vlag", 
                   center=0, 
                   yticklabels=False)
    
    #ax.set(xlabel="k in k-mean clustering", ylabel="Sample")
    plt.savefig("../../figures/figures_from_template/heatmap_sc3s_2000rep_countSampleInCluster_"+compound+".svg")

    #Calculate percentage of each cluster which occur in 9 of 10 replicates 
    data_percent = pd.DataFrame.from_dict({'sc3s_100': [0.0]*10,
                                           'sc3s_110': [0.0]*10,
                                           'sc3s_120': [0.0]*10,
                                           'sc3s_150': [0.0]*10,
                                           'sc3s_180': [0.0]*10
                                          })
    data_clusterSize = pd.DataFrame.from_dict({'sc3s_100': [0.0]*10,
                                           'sc3s_110': [0.0]*10,
                                           'sc3s_120': [0.0]*10,
                                           'sc3s_150': [0.0]*10,
                                           'sc3s_180': [0.0]*10
                                          })
    
    for k in range(10):
        data_temp = pd.read_parquet('/share/data/analyses/camilla/cp_tpp/processed_data/pca_specs_all_withClusters_2000runs'+str(k)+'.parquet')
        for column in data_repeat.columns:
            cluster_no = list(data_temp[data_temp["cbkid"] == compound_dict[compound]][column])[0]
            sample_total_cluster = 0.0
            for i in data_repeat.index:
                if data_temp.loc[i,column] == cluster_no:
                    sample_total_cluster += 1.0
                    #Check if sample i occur in 9 of 10 replicates
                    if data_repeat.loc[i,column] >= 8:
                        data_percent.loc[k,column] += 1.0
            data_percent.loc[k,column] = data_percent.loc[k,column]/sample_total_cluster*100
            data_clusterSize.loc[k,column] = sample_total_cluster

    #Save tables for data_percent and data_clusterSize
    data_percent.to_latex("../../tables/sc3s_reproducibility_percentSampleIn8of10Clusters_2000rep_"+compound+".txt")
    data_clusterSize.to_latex("../../tables/sc3s_reproducibility_clusterSize_2000rep_"+compound+".txt")
```
